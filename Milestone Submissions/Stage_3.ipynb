```python
import os

# From our colab notebook
#from google.colab import drive
#drive.mount("/content/drive")
#dir = '/content/drive/MyDrive/4152 5152 Group Work/data/all'

# Inside the gitlab repo
# ./all_data for windows
# /all_data for mac/linux (unconfirmed)
dir = './all_data'
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import random
import cv2

cols = 5
rows = 6

random_images = random.choices(os.listdir(dir), k=30)

fig, axs = plt.subplots(rows, cols, figsize=(20,8))


for i, image_file in enumerate(random_images):
  row = i // cols
  col =  i % cols
  image_path = os.path.join(dir, image_file)
  image = cv2.imread(image_path)
  axs[row, col].imshow(image)
  axs[row, col].set_title(image_file)
  axs[row, col].axis('off')
```
